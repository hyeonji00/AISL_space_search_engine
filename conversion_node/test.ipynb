```python
with open('./sejong_node.json', 'r') as file:
    data = json.load(file)

def mul_ten(x):
    return x*10

scaled_data = list(map(mul_ten, data["node_locations"]))
nested_list = list()
for i in data["node_locations"]:
    nested_list.append(list(map(mul_ten, i)))

print(nested_list)

data["node_locations"] = nested_list
data = json.dumps(data)
with open('./sejong_node_mean_scaled_ten.json', 'w') as file:
    file.write(data)
```
